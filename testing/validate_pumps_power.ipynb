```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_simulink = pd.read_csv('Pumps_P_Simulink.csv')
df_pyfmi=pd.read_csv('Pumps_P_Pyfmi.csv')
```

```python
plt.figure(figsize=(24, 16))
plt.plot(range(0,10000),df_simulink['Pcdwp_sum'].iloc[1:10001], label='Simulink')
plt.plot(range(0,10000),df_pyfmi['P_CDWPs_sum'].iloc[0:10000]/1000, label='Pyfmi')
plt.legend(fontsize=25)
plt.title('Simulink vs. Pyfmi P_CDWPs_sum Comparative Analysis',fontsize=30)
plt.xlabel('Time step', fontsize=20)
plt.ylabel('P_CDWPs_sum', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(24, 16))
plt.plot(range(0,10000),df_simulink['Pchwp_sum'].iloc[1:10001], label='Simulink')
plt.plot(range(0,10000),df_pyfmi['P_CHWPs_sum'].iloc[0:10000]/1000, label='Pyfmi')
plt.legend(fontsize=25)
plt.title('Simulink vs. Pyfmi P_CHWPs_sum Comparative Analysis',fontsize=30)
plt.xlabel('Time step', fontsize=20)
plt.ylabel('P_CHWPs_sum', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(24, 16))
plt.plot(range(8000,8000+200),df_simulink['Pcdwp_sum'].iloc[8000+1:8000+200+1], label='Simulink')
plt.plot(range(8000,8000+200),df_pyfmi['P_CDWPs_sum'].iloc[8000:8000+200]/1000, label='Pyfmi')
plt.legend(fontsize=25)
plt.title('Simulink vs. Pyfmi P_CDWPs_sum Comparative Analysis',fontsize=30)
plt.xlabel('Time step', fontsize=20)  # X轴标签及其字体大小
plt.ylabel('P_CDWPs_sum', fontsize=20)  # Y轴标签及其字体大小
plt.xticks(fontsize=15)  # X轴刻度标签的字体大小
plt.yticks(fontsize=15)  # Y轴刻度标签的字体大小
plt.show()
```
